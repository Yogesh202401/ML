### Step 1 :Read dataset

```python
import pandas as pd
df = pd.read_csv("iris (1).csv")
df.head()
```

### Step 2: Perform basic data quality checks


```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

### Step 3: Seperate X and Y(species)


```python

X = df.drop(columns=['species'])
Y = df[['species']]
```

```python
X.head()
```

```python
Y.head()
```

### Step 4: Perform preprocessing on X


```python
X.dtypes
```

```python

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python

X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

### Step 5: Perform train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.33, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
df['petal_length'] = df['petal_length'].astype(float)
```

### Step 6 : Build the base model

```python

from sklearn.tree import DecisionTreeClassifier
model1 = DecisionTreeClassifier(max_depth=1,
                                min_samples_split=5,
                                min_samples_leaf=5,
                                criterion='entropy')
model1.fit(xtrain, ytrain)
```

```python
model1.score(xtrain,ytrain)
```

```python
model1.score(xtest,ytest)
```

### Step 7: Hyperparameter Tuning - GridSearchCV


```python
params = {'max_depth':[1, 2, 3, 4, 5],
          'min_samples_split':[2, 3, 4, 5, 6, 7],
          'min_samples_leaf':[2, 3, 4, 5, 6, 7],
          'criterion':['gini', 'entropy']}
```

```python
from sklearn.model_selection import GridSearchCV
dtc = DecisionTreeClassifier()
gscv = GridSearchCV(dtc, param_grid=params, cv=5, scoring='f1_macro')
gscv.fit(xtrain, ytrain)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
best_dtc = gscv.best_estimator_
best_dtc
```

### Model Evalution

```python
# confusion matrix
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(best_dtc, xtrain, ytrain)
```

```python
ConfusionMatrixDisplay.from_estimator(best_dtc, xtest, ytest)
```

```python
# Predict results for train and test
ypred_train = best_dtc.predict(xtrain)
ypred_test = best_dtc.predict(xtest)
```
